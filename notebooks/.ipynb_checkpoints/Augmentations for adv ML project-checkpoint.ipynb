#### Implement a set of "Joint Transform" functions to perform data augmentation in your dataset. 

Neural networks are typically applied to transformed images. There are several important reasons for this: 

1. The image data should is in certain required format (i.e. consistent spacial resolution to batch). The images should also be normalized and converted to the "tensor" data format expected by pytorch libraries. 

2. Some transforms are used to perform randomized image domain transformations with the purpose of "data augmentation". 

In this exercise, you will implement a set of different transform functions to do both of these things. Note that unlike classification nets, training semantic segmentation networks requires that some of the transforms are applied to both image and the corresponding "target" (Ground Truth segmentation mask). We refer to such transforms and their compositions as "Joint". In general, your Transform classes should take as the input both the image and the target, and return a tuple of the transformed input image and target. Be sure to use critical thinking to determine if you can apply the same transform function to both the input and the output.

For this problem you may use any of the `torchvision.transforms.functional` functions. For inspiration, refer to:

#### (a) Implement RandomFlip

This class should take a img, target pair and then apply a horizontal flip across the vertical axis at random.

#### Solution:

```python
class randomFlip(object):
    def __init__(self, prob):
        """
        params:
            prob (float) : the probability of flipping the image 
        """
        self.prob = prob
        
    def __call__(self, img, target):
        decision = np.random.choice([0,1],p=[self.prob,1-self.prob])
        if decision == 0:
            return (tF.hflip(img), 
                tF.hflip(target))
        else:
            
            return (img,target)
    
img, target = randomFlip(1)(*sample1)
print(np.array(img).shape)
fig = plt.figure(figsize=(12,6))
ax1 = fig.add_subplot(2,2,1)
ax1.imshow(img)
ax2 = fig.add_subplot(2,2,2)
ax2.imshow(target)
```

#### (b) Implement RandomResizeCrop

This class should take a img, target pair and then resize the images by a random scale between $[\mbox{minimum_scale}, \mbox{maximum_scale}]$, crop a random location of the image by $\min(\mbox{size}, \mbox{image_height}, \mbox{image_width})$ (where the size is passed in as an integer in the constructor), and then resize to $\mbox{size} \times \mbox{size}$ (again, the size passed in). The crop box should fit within the image.

#### Solution:

```python
class randomResizeCrop(object):
    def __init__(self,minScale,maxScale,size):
        """
        params:
            minScale (float) : minimum amount by which to scale the image
            maxScale (float) : maximum amount by which to scale the image
            size (int) : dimensions of the final image 
        """
        self.maxScale = maxScale
        self.minScale = minScale
        self.size = size
        
        if self.minScale < 1:
            self.minScale = 1
            print("minScale must be at least 1")
            
        if self.maxScale <= self.minScale:
            raise ValueError("maxScale {} must be greater than minScale {}".format(self.maxScale,self.minScale))
        
    def __call__(self, img, target):
        w,h = img.size
        scale = np.random.rand() * (self.maxScale - self.minScale) + self.minScale
        #random scale
        img = tF.affine(img,scale=scale,angle=0,translate=[0,0],shear=0)
        target = tF.affine(target,scale=scale,angle=0,translate=[0,0],shear=0)
        w,h = img.size
        dim = np.min([self.size,h,w])
        top=np.random.randint(0,h-dim)
        left=np.random.randint(0,w-dim)
        #random crop
        img = tF.crop(img,top=top,left=left,height=dim,width=dim)
        target = tF.crop(target,top=top,left=left,height=dim,width=dim)
        #final resize
        return (tF.resize(img,(self.size,self.size)),
               tF.resize(target,(self.size,self.size)))
        
        
for x in range(5): #loop to help check for bugs
    img, target = randomResizeCrop(minScale=1.2,maxScale=2,size=350)(*sample1)
    fig = plt.figure(figsize=(12,6))
    ax1 = fig.add_subplot(2,2,1)
    ax1.imshow(img)
    ax2 = fig.add_subplot(2,2,2)
    ax2.imshow(target)


```

#### (c) Implement Normalize

This class should take a img, target pair and then normalize the images by subtracting the mean and dividing variance. 

#### Solution:

```python
class Normalize(object):
    def __init__(self,mean,variance):
        """
        params:
            minScale (float) : minimum amount by which to scale the image
            maxScale (float) : maximum amount by which to scale the image
            size (int) : dimensions of the final image 
        """
        self.mean =  torch.tensor(mean)#[np.newaxis,np.newaxis,:]
        self.variance =  torch.tensor(variance)#[np.newaxis,np.newaxis,:]
        
    def __call__(self, img, target):
        if type(img) != torch.Tensor:
            print('not tensor')
            img = torch.from_numpy(np.array(img)).type(torch.float32)
        
        if len(img.shape) != 3:
            print("Normalize only implemented for color images, input image had shape {}".format(img.shape))
            raise NotImplementedException
            
        if np.min(img.shape) != 3:
            print("Normalize only implemented for RGB images, input image had {} channels".format(np.min(img.shape)))
            raise NotImplementedException
            
        
                  
        if img.shape[2] == 3:
            img = img.permute(-1, 0, 1)
                  
        img = (img - self.mean[:, None, None]) / self.variance[:, None, None]
#         img = img.
        assert(img.shape[0] == 3)
        return (img,target)
        
        
norm = ([0.485, 0.456, 0.406], 
        [0.229, 0.224, 0.225]) 

img, target = Normalize(*norm)(*sample1)
print(img.shape)
fig = plt.figure(figsize=(12,6))
ax1 = fig.add_subplot(2,2,1)
ax1.imshow(img.permute(1,2,0).numpy())
ax2 = fig.add_subplot(2,2,2)
ax2.imshow(target)
```

#### (d) Compose the transforms together:

```python
# This class composes transofrmations from a given list of image transforms (expected in the argument). Such compositions 
# will be applied to the dataset during training. This cell is fully implemented.

class JointCompose(object):
    def __init__(self, transforms):
        """
        params: 
           transforms (list) : list of transforms
        """
        self.transforms = transforms

    # We override the __call__ function such that this class can be
    # called as a function i.e. JointCompose(transforms)(img, target)
    # Such classes are known as "functors"
    def __call__(self, img, target):
        """
        params:
            img (PIL.Image)    : input image
            target (PIL.Image) : ground truth label 
        """
        assert img.size == target.size
        for t in self.transforms:
            img, target = t(img, target)
        return img, target
```

#### (a) Implement encoder/decoder segmentation CNN using PyTorch.

You must follow the general network architecture specified in the image below. Note that since convolutional layers are the main building blocks in common network architectures for image analysis, the corresponding blocks are typically unlabeled in the network diagrams. The network should have 5 (pre-trained) convolutional layers (residual blocks) from "resnet" in the encoder part, two upsampling layers, and one skip connection. For the layer before the final upsampling layer, lightly experiment with some combination of Conv, ReLU, BatchNorm, and/or other layers to see how it affects performance.

#### Solution:


goal size ---> torch.Size([1, 3, 375, 500])
intermediate goal size ---> torch.Size([1, 64, 94, 125])
output size ---> torch.Size([1, 512, 12, 16])

```python
import torchvision.models as models

class MyNet(nn.Module):
    def __init__(self, num_classes, criterion=None,h=375,w=500,verbose=False):
        super(MyNet, self).__init__()
        self.verbose=verbose
        self.num_classes = num_classes
        self.criterion = criterion
        self.resnet = models.resnet18(pretrained=True)
        
        self.encoder1 = nn.Sequential(*[self.resnet._modules[x] for x in ['conv1',
                'bn1','relu','maxpool','layer1']]
                )
                
        self.encoder2 = nn.Sequential(*[self.resnet._modules[x] for x in ['layer2','layer3','layer4']]
                )
        
        with torch.no_grad():
            out1 =  self.encoder1(torch.randn(1,3,h,w))
            _, C1, H1, W1 = out1.shape
            out2 = self.encoder2(out1)
            _, C2, H2, W2 = out2.shape

        
        self.upsample1 = nn.Sequential(
            nn.Upsample(size=(H1,W1), mode='bilinear', align_corners=True)
        )
                
        self.someConvsRelusBNs = nn.Sequential(
            torch.nn.Conv2d(in_channels=C2+C1, out_channels=128, kernel_size=3, stride=1, padding=1, bias=False),
            nn.BatchNorm2d(128, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),
            nn.ReLU(inplace=True),
            torch.nn.Conv2d(in_channels=128, out_channels=32, kernel_size=3, stride=1, padding=1, bias=False),
            nn.BatchNorm2d(32, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),
            nn.ReLU(inplace=True),
            torch.nn.Conv2d(in_channels=32, out_channels=self.num_classes, kernel_size=3, stride=1, padding=1, bias=False),
            nn.BatchNorm2d(self.num_classes, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),
            nn.ReLU(inplace=True),
            nn.Conv2d(in_channels=self.num_classes, out_channels=self.num_classes, kernel_size=1)
        )
        
        self.upsample2 = nn.Sequential(
            nn.Upsample(size=(h,w), mode='bilinear', align_corners=True)
        )
        
        
    def predict(self,inp,dynUpsample=False):
        """Method to generate a prediciton from the network"""
        skip = self.encoder1(inp)
        if self.verbose: print("encoder1:",skip.shape)
        x = self.encoder2(skip)
        if self.verbose: print("encoder2:",x.shape)
            
        if dynUpsample:
            _, C, H, W = skip.shape
            x = nn.Upsample((H,W), mode='bilinear', align_corners=True)(x)
        else:
            x = self.upsample1(x)
            
        if self.verbose: print("upsample1:",x.shape)
        x = torch.cat([skip,x],dim=1)
        if self.verbose: print("concat:",x.shape)
        x = self.someConvsRelusBNs(x)
        if self.verbose: print("someConvsRelusBNs:",x.shape)
            
        if dynUpsample:
            _, C, H, W = inp.shape
            x = nn.Upsample((H,W), mode='bilinear', align_corners=True)(x)
        else:
            x = self.upsample2(x)
        
        if self.verbose: print("upsample2:",x.shape)
        
        return x
    
    def forward(self, inp, gts=None, dynUpsample=False):
        x = self.predict(inp,dynUpsample=dynUpsample)
        
        if gts != None:
            return x, self.criterion(x, gts)
        else: 
            return x, None
        
    

```

#### (a) Implement the loss function (Cross Entropy Loss) as detailed below. For debugging, part (b) below allows to compare the values w.r.t. the standard function "nn.CrossEntropyLoss". If your loss function works, uncomment the use of "MyCrossEntropyLossy" replacing "nn.CrossEntropyLoss" everywhere below in the notebook (just a couple of places).

You should return the mean (average over all batch pixels) negative log of soft-max for the ground truth class. Assuming that  
$X_p^m$ are logits at pixel $p$ for $C$ classes, so that $m\in\{0,1,...,C-1\}$, each point $p$ contributes to this loss $$ -\log\frac{e^{X_p^{y_p}}}{\sum_m e^{X_p^m}}\;\;=\;\;-\sum_k Y_p^k \log\frac{e^{X_p^k}}{\sum_m e^{X_p^m}} \quad\quad\quad\quad(*)$$ where $y_p$ is the ground truth label and $Y_p^k$ is the corresponding one-hot distribution. You should implement the basic math, as in one of the equations above. 

NOTE 1: Numerically robust implementation of soft-max is not immediately straightforward. Indeed, logits $X_p^k$ are arbitrary numbers and their exponents could be astronomically large. While the log cancels out the exponent in the enumerator (do it mathematically, not numerically), the sum of exponents in the denominator can not be easily simplified and this is where the numerical issues hide. A naive implementation adding huge values of logits' exponents easily loses precision. Instead, one should use the right hand side in the following algebraically equivalent formulation: $$\log (e^A + e^B + \dots + e^Z) \;\;\equiv\;\; \mu + \log (e^{A-\mu} + e^{B-\mu} + \dots + e^{Z-\mu})$$ where $\mu := \max\{A,B,\dots,Z\}$. The right hand side is numerically stable since the arguments of the exponents are negative and the exponents are all bounded by value $1$.


HINT 1: Similarly to many previous assignemnts, avoid for-loops. In fact, you should not use none below. Instead, use standard  opertors or functions for tensors/matrices (e.g. pointwise addition, multiplication, etc.) In particular, you will find useful functions like "sum" and "max" that could be applied along any specified dimension of the tensor. Of course, torch also has pointwise functions "log" and "exp" that could be applied to any tensor.

HINT 2: Be careful - you will have to work with tensors of different shapes, e.g. even input tensors "targets" and "logits" have different shapes. As in previous assignments based on "numpy", you should pay attention to tensors' dimensions in pyTorch. For example, pointwise addition or multiplication requires either the same shape or shapes "brodcastable" to the same shape (similar in "pyTorch" and "numpy", e.g. see <a href="https://numpy.org/doc/stable/user/theory.broadcasting.html#array-broadcasting-in-numpy">here</a>). If in doubt, use $$print(xxx.size())$$ to check the shape - I always do! If needed, modify the shapes using standard functions: "transpose" to change the order of the dimentions, "reshape", "flatten", or "squeeze" to remove dimensions, "unsqueeze" to add dimentions. You can use other standard ways to modify the shape, or rely on broadcasting.

HINT 3: Your loss should be averaged only over pixels that have non-void labels. That is, exclude pixels with "ignore index" labels. For example, you can compute a "mask" of non-void pixels, and use it to trim non-void pixels in both the "targets" and "logits". You might get an inspiration from the "mask" in your K-means implementation. 

HINT 4: For simplicity, you may "flatten" all tensors' dimensions corresponding to batches and image width & height. The loss computation does not depend on any information in these dimentions and all (non-void) pixels contribute independently based on their logits and ground truth labels (as in the formula above). 

HINT 5: In case you want to use the right-hand-size in $(*)$, you can use the function "torch.nn.functional.one_hot" to convert a tensor of "target" class labels (integers) to the tesnor of one-hot distributions. Note that this adds one more dimension to the tensor. In case you want to implement the left-hand-side of $(*)$, you can use the function "torch.gather" to select the ground truth class logits $X_p^{y_p}$. 

HINT 6: Just as some guidence, a good solution should be around ten lines of "basic" code, or less.

```python
# Student Answer:
class MyCrossEntropyLoss(object):
    
    def __init__(self, ignore_index=255):
        self.ignore_index = ignore_index
        
    def __call__(self, logits, targets):
        # N - batch size, C - number of classes (excluding void), HxW - image size
        N, C, H, W = logits.size()
        #stable softmax
        max_smx = torch.max(logits,dim=1,keepdims=True).values
        womax = logits - max_smx
        exp_on_womax = torch.exp(womax)
        smx_over_logits = logits - (max_smx + torch.log(exp_on_womax.sum(dim=1,keepdims=True)))
        n,h,w = torch.where(targets != self.ignore_index)
        targets = targets[n,h,w]
        oneHotTargets = nn.functional.one_hot(targets,num_classes=C)
        indexIntoSmx = smx_over_logits[n,:,h,w]
    
        return (-1 * (indexIntoSmx * oneHotTargets).sum(dim=1,keepdims=True)).mean()
    
```

#### (a) Define an optimizer to train the given loss function.

Feel free to choose your optimizer of choice from https://pytorch.org/docs/stable/optim.html.


```python
def get_optimizer(net,lr):
    optimizer = torch.optim.SGD(
        net.parameters(), lr=lr
    )
    return optimizer
```




#### (b) Write the training loop to train the network. 


```python
def train(train_loader, net, optimizer, loss_graph, dynUpsample=False):
    iou_arr = []
    loss_graph = []
    for i, data in enumerate(train_loader):
        inputs, masks = data

        if USE_GPU:
            inputs, masks = inputs.to('cuda'), masks.to('cuda')
            
        x, loss  = net(inputs,masks,dynUpsample=dynUpsample)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
            
        with torch.no_grad():
            preds = x.max(dim=1).indices.cpu().numpy()
            gts = masks[None].cpu().squeeze(0).numpy()

            conf = eval_semantic_segmentation(preds, gts)

            iou_arr.append(conf['miou'])
            loss_graph.append(loss.detach().item()) #Populate this list to graph the loss

    return np.mean(loss_graph), np.mean(iou_arr)
```

Firstly, we see that the network trained on the entire dataset generalizes much better then that overfit network: 0.28 vs 0.20 IOU, an 8% improvement. This may seem modest, but since we are using an architecture which was not thoroughtly explored, we can not expect to acheive exceptional performance even when training on the entire dataset. 

### Potential improvements and Obvious issues

While the network trained in this notebook acheives moderate performance, many improvements to its architecture and the training scheme used could lead to improved overall performance. 

As it is, the network architecture utilizes agressive upsampling layers to acheive the desired end resolution (obvious issue 1). Although some upsampling is necessary in almost any segmentation network (needed for generalizing to arbitrary spatial resolutions), it may be helpfull to use a less agressive upsampling scheme. Transposed convolutions are a usefull tool for parameteric upsampling, allowing the network learn an efficient upsampling process, which may be valuable for segmenting minute details. For instance, we could modify the current network architecture to first upscale the image using transposed convolutions before finally using a minimal upscale to obtain the desired final resolution. Another potential solution to this problem would be to use an encorder which does not reduce the spatial resolution as much, this way the upsampling layers would not be responsible for such a large upsampling. Aditionally, further improvements could be realized by tuning the architecture structure (e.g. convs, batch norms, relus) based on out of sample performance assessment(validation set performance). Altogether, there are many avenues to explore for improving the current network architecture, yet, personally, the most obvious is to reduce the amount of upsampling done exclusively by upsapling layers


Another way to improve performance is to modify the training scheme/dataset, specifically the training is currently limited due to the small crop size of the images in the training set (obvious issue 2). It stands to reason that using larger images might improve performance, however, I chose to use small images in order to use the 'randomScaleandCrop' with all the training images (requires cropping to the minimum image size). To improve perfromance, we could try to resize training images to obtain larger images for input. This would make the training targets more similar in size to the validation images. Other improvements could come from using an appropriate learning rate scheduler, or futher tuning the hyperparameters of the optimizer.
